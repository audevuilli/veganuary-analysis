First open the different files where the tweets have been saved and create a dataframe with the parameters of the tweets we are interested in it!

```python
tweet_file = ['#Veganuary2019_2019-01-22_to_2019-01-28.json',
              '#Veganuary2019_2019-01-25_to_2019-01-31.json',
              '#Veganuary2019_2019-01-31.json',
              '#veganuary_2019-01-25_to_2019-01-31.json',
              '#veganuary_2019-01-31.json']
```

```python
# Read all the files and append them to a list!
tweets = []
for file in tweet_file:
    with open(file, 'r') as f:
        for line in f.readlines():
            tweets.append(json.loads(line))
```

```python
# Read the data into a dataframe
from pandas.io.json import json_normalize
df = pd.DataFrame.from_dict(json_normalize(tweets), orient='columns')
```

```python
# Drop the duplicated data and save the data into a csv file. 
df = df.drop_duplicates('created_at')
```

# Analysis of Sentiment - Data Cleaning


```python
# For Sentiment Analysis
import string
import nltk

from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
import sklearn
import gensim
```

```python
#import nltk and download all stopwords and punctuations
nltk.download('stopwords')
nltk.download('punkt')

import re
```

```python
twittertext = df
```

```python
twittertext.head(2)
```

```python
# Remove all the https
twittertext.Text = twittertext.Text.str.replace(r'http\S+','')
```

```python
# Remove all the '\n' line-brakes
twittertext.Text = twittertext.Text.str.replace(r'\n','')
```

```python
# Remove 'non-ascii' characters
twittertext.Text = twittertext.Text.str.replace(r'[^\x00-\x7F]','')
# Make all the text lowercase
twittertext.Text = twittertext.Text.str.lower()
```

```python
#create a column for all #hasthtags that appears in the text
twittertext['hashtags'] = twittertext.Text.str.findall(r'#\S+')
#create a column for all @twitter handles that appears in the tweets
twittertext['twitter_handles'] = twittertext.Text.str.findall(r'@[\w]*')
```

```python
# Remove the #hashtags and the twitter_handles from the tweets
#twittertext.Text  = twittertext.Text.str.replace(r'#\S+','')
# Remove the @twitter_handles from the tweets
twittertext.Text  = twittertext.Text.str.replace(r'@[\w]*','')
```

```python
# Remove the particular character 
twittertext['cleanText'] = twittertext.Text.str.replace(r'[/.!?$%^&*()#:@]','')
```

```python
twittertext['cleanText'] = twittertext.Text.apply(lambda x:x.translate(str.maketrans('','', string.punctuation)))
```

```python
#tokenise all of the clean text
twittertext['tokenText'] = twittertext['cleanText'].apply(word_tokenize)
```

```python
twittertext['tokenCount'] = twittertext.tokenText.apply(len)
```

```python
twittertext.head(2)
```

```python
# Import stopwords with nltk.
stop = stopwords.words('english')

punctuation = list(string.punctuation)
other = ['rt', '3', '25', '5', '6', '7', '8', '9', ':/', '10', '4', '1', 
         '2', 'us', 'It\'s', 'via', '‡§æ', '', 'RT', '‚Ä¶', 'I', '‚Äô', 'The', 
         'what', 'What', 'You', 'Your', 'A', 'https', 'get', 'amp','im',
         'tweet','retweet','could']

stop = stop + punctuation + other


# Exclude stopwords with Python's list comprehension and pandas.DataFrame.apply.
twittertext['noStopWords'] = twittertext['cleanText'].apply(
    lambda x:' '.join([word for word in x.split() if word not in (stop)]))
```

```python
#tokenise all of the text with Stop words
twittertext['tokenTextnoStop'] = twittertext.noStopWords.apply(word_tokenize)
```

### Understand the top 20 words in the Tweets

```python
from collections import Counter
 
count_all = Counter()
for line in twittertext['tokenTextnoStop']:
    # Create a list with all the terms
    terms_all = [term for term in line]
    # Update the counter
    count_all.update(terms_all)
    # Print the first 5 most frequent words
print(count_all.most_common(20))
```

Here we need to bear in mind that many of the tweets are actually retweeted tweet. We need therefore to delete the duplicate tweet. We can do this by droping the duplicate from clean Text column. 

```python
twittertext_nodup = twittertext.drop_duplicates('Text')
```

```python
from collections import Counter
 
count_all = Counter()
for line in twittertext_nodup['tokenTextnoStop']:
    # Create a list with all the terms
    terms_all = [term for term in line]
    # Update the counter
    count_all.update(terms_all)
    # Print the first 5 most frequent words
print(count_all.most_common(20))
```

```python
most_common = pd.DataFrame(count_all.most_common(20))
most_common['Freq'] = (most_common[1]/8686)*100
most_common
```

```python
data = [Bar(
        x=most_common[0], # assign x as the dataframe column 'x'
        y=most_common['Freq']
    )]

layout = Layout(
    title="Top 10 most used words in Tweets",
    xaxis=dict(title='Most famous words'),
    yaxis=dict(title='Percentage of word appearance in the tweets'),
    width = 700
)

fig = Figure(data=data, layout=layout)

iplot(fig)
```
